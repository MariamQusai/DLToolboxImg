## horizontal and vertical random flips

```python
def random_flip_img(X,Y, horizontal_chance=0, vertical_chance=0):
    import cv2
    flip_horizontal = False
    if random.random() < horizontal_chance:
        flip_horizontal = True

    flip_vertical = False
    if random.random() < vertical_chance:
        flip_vertical = True

    if not flip_horizontal and not flip_vertical:
        return (X,Y)

    flip_val = 1
    if flip_vertical:
        flip_val = -1 if flip_horizontal else 0

    Xaug = np.array([cv2.flip(x, flip_val) for x in X])
    Yaug = np.array([cv2.flip(x, flip_val) for x in Y])

    return (Xaug,Yaug)

```

```python
h=0
v=1
Xaug,Yaug=random_flip_img(X,Y, horizontal_chance=h, vertical_chance=v)
```
